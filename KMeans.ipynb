K-Means Cluster

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("/content/Cust_Segmentation.csv")
data.drop(columns="Address")
data
```

```python
data = pd.read_csv("/content/Cust_Segmentation.csv")
x =data.drop(columns="Address")
print(x)
```

```python
model =KMeans(init="random",n_clusters=3)

model.fit(x.dropna())
```

```python
centroid =model.cluster_centers_
print("centroid ",centroid )
label =model.labels_
print("labels",label)
```

```python
plt.scatter(data[: ,0] ,data[: ,1],c="green", cmap='viridis', marker='o', s=50, label='Data Points')
plt.scatter(centroid[:, 0], centroid[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.show()
```

Iris Dataset

```python
# Using Iris dataset for this clustering Algorithm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data =pd.read_csv("/content/IRIS.csv")

dt =data.drop(columns="species")

dt
```

```python
model =KMeans(init="random",n_clusters=3)

model.fit(dt)

center = model.cluster_centers_
label = model.labels_

print("centroid ",center)
print("labels" ,label)
print("inertia" ,model.inertia_)
```

# simple example for KMeans

```python
# simple example for KMeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x =[2,4,6,8,2 ,4,7,2,4,3]
y =[5,4,3,2,8 ,4,3,1,8,6]

data =np.array([x,y])

plt.scatter(x,y)
```

```python
model = KMeans(n_clusters=2,init="k-means++")

model.fit(data)

center =model.cluster_centers_
label = model.labels_

print(center)
print(label)
```

```python
plt.scatter(data[: ,0],data[: ,1] , c="green")
plt.scatter(center[: ,0],center[ : ,1 ] ,c="red" ,marker="*")
plt.show()
```
