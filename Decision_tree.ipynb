Decision Tree

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

dt ={"age":[25,26,27,35,12,22,50,42,38,46],
     "salary":[25000,26000,27000,55000,12000,22000,50000,62000,58000,76000],
     "order":["no","no","no","yes","no","no","yes","yes","yes","yes"]}

data =pd.DataFrame(dt)

x = data[["age","salary"]]
y = data["order"]

model = DecisionTreeClassifier(criterion='entropy', max_depth=4,random_state=42 )

model.fit(x,y)

test =pd.DataFrame({"age":[34],"salary":[10000]})
pred =model.predict(test)

print(pred)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier ,plot_tree
from sklearn.model_selection import train_test_split

dt ={"age":[25,26,27,35,12,22,50,42,38,46],
     "salary":[25000,26000,27000,55000,12000,22000,50000,62000,58000,76000],
     "order":["no","no","no","yes","no","no","yes","yes","yes","yes"]}

data =pd.DataFrame(dt)

x = data[["age","salary"]]
y = data["order"]

x1,x2 ,y1,y2 = train_test_split(x,y)

model = DecisionTreeClassifier(criterion='gini', max_depth=4,random_state=42 )

model.fit(x1,y1)

pred =model.predict(x2)

print(x2,"->",pred)

plt.figure(figsize=(10,8))
plot_tree(model,feature_names=data.columns, class_names=model.classes_, filled=True)
plt.show
```

```python
# Loan Appruval (decision Tree)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier ,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data =pd.read_csv("/content/bankloan.csv")

print(data.head(10))

x =data[["ID","Age","Experience","Income","ZIP.Code","Family","Education","Mortgage"]]
y =data["Personal.Loan"]

x_train,x_test ,y_train,y_test =train_test_split(x,y)

model =DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=42)

model.fit(x_train,y_train)

y_pred =model.predict(x_test)
print("predictions" ,y_pred)

acc =accuracy_score(y_test,y_pred)
print("accuracy" ,acc)

plt.figure(figsize=(10,10))
plot_tree(model)
plt.show
```

Random Forest Model

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,r2_score

dt=pd.read_csv("/content/BostonHousing.csv")
data =dt.head(100)
print(data)
```

```python
x1 =data.drop(columns="medv")
y1 =data["medv"].dropna()

x1,x2, y1,y2 = train_test_split(x1,y1)

model =RandomForestRegressor(max_depth=10,random_state=42)

model.fit(x1,y1)

y3 = model.predict(x2)
print(y3)
print("R2 score",r2_score(y2,y3))
```
