Principle Component Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition  import PCA

data =[[50,150],
       [60,160],
       [70,170],
       [80,180]]

dt =StandardScaler()
x = dt.fit_transform(data)


model = PCA(n_components=1)

x1 = model.fit_transform(x)

print("Standard ",x,"\nReduced",x1)


plt.scatter(x[:,0],x[:,1],c="r")
```

```python
plt.scatter(x[:,0],x[:,0],c="r")
plt.scatter(x1[:,0],x1[:,0],c="g")
```

PCA Using Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

dt = pd.read_csv("/content/IRIS.csv")

x =dt.drop(columns="species")
y =dt["species"]

scaler = StandardScaler()
x_std = scaler.fit_transform(x)

model = PCA(n_components=2)

x1 =model.fit_transform(x_std)

plt.scatter( x1[:,0] ,x1[:,1] ,c="r")

```
