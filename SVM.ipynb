```python
# people age to life stage SVM

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC

x = np.array([[5],[4],[7],[11],[15], [18],[20],[19],[25],[29], [35],[30],[38],[40],[45]])
y = np.array(["child","child","child","child","child", "boy","boy","boy","boy","boy", "men","men","men","men","men"])

model = SVC(kernel ="linear", degree=3, gamma="scale")

model.fit(x,y)

x2 =np.array([[8],[23],[42]])
y2 =model.predict(x2)
print(y2)

plt.scatter(x,y ,color="blue" ,label="train")
plt.scatter(x2,y2 ,color="green" ,label="test")
plt.legend
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Data: Ages and their corresponding labels
x = np.array([[5], [4], [3], [7], [11], [15], [18], [20], [19], [25], [29], [35], [30], [38], [40], [45]])
y = np.array(["child", "child", "child", "child", "boy", "boy", "boy", "boy", "boy",
              "men", "men", "men", "men", "men", "men", "men"])

# Create and fit SVM model
model = SVC(kernel="linear", gamma="scale")
model.fit(x, y)

# Predict the age group for age 28
predicted_class = model.predict([[28]])
print(f"The age group for age 28 is: {predicted_class[0]}")

# Plotting the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plotting the decision boundary (hyperplane)
# The decision function gives us the distance from the hyperplane
x_range = np.linspace(0, 50, 500).reshape(-1, 1)
decision_function = model.decision_function(x_range)

# SVM makes a decision at the point where decision_function = 0 (the hyperplane)
plt.plot(x_range, model.predict(x_range), color='red', linestyle='--', label='Decision Boundary (Hyperplane)')

# Highlight the predicted point
plt.scatter([28], predicted_class, color='green', s=100, label='Predicted Point (28 years)')

# Highlighting support vectors
plt.scatter(model.support_vectors_, model.predict(model.support_vectors_),
            color='black', s=150, facecolors='none', edgecolors='k', label='Support Vectors')

# Labels and Title
plt.xlabel('Age')
plt.ylabel('Category')
plt.title('SVM Classification with Hyperplane')
plt.legend()
plt.show()
```
